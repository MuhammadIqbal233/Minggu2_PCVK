```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('gradient (1).jpg', 0)
# Binary threshold
_, binary = cv2.threshold(img, 170, 255, cv2.THRESH_BINARY)

# Inverse binary threshold
_, binary_inv = cv2.threshold(img, 170, 255, cv2.THRESH_BINARY_INV)

# Truncate threshold
_, truncate = cv2.threshold(img, 170, 255, cv2.THRESH_TRUNC)

# Threshold to zero
_, tozero = cv2.threshold(img, 170, 255, cv2.THRESH_TOZERO)

# Inverse threshold to zero
_, tozero_inv = cv2.threshold(img, 170, 255, cv2.THRESH_TOZERO_INV)
titles = ['Original Image', 'Binary', 'Inverse Binary', 'Truncate', 'To Zero', 'Inverse To Zero']
images = [img, binary, binary_inv, truncate, tozero, tozero_inv]

for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('balloon.jpg', 0)

# Apply global thresholding with threshold value of 90
ret, thresh_global = cv2.threshold(img, 90, 255, cv2.THRESH_BINARY)

# Apply Otsu thresholding
ret, thresh_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display the original image and the thresholded images
plt.subplot(131), plt.imshow(img, cmap='gray')
plt.title('Citra Asli'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(thresh_global, cmap='gray')
plt.title('Global Thresholding (v=90)'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(thresh_otsu, cmap='gray')
plt.title('Otsu Thresholding(v=186)'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Import necessary libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image
img = cv2.imread('peppers.jpg')

# Convert the image to HSV color space
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Define the range of yellow color in HSV
lower_yellow = np.array([20, 100, 100])
upper_yellow = np.array([30, 255, 255])

# Threshold the HSV image to get only yellow colors
mask = cv2.inRange(hsv, lower_yellow, upper_yellow)

# Bitwise-AND mask and original image
res = cv2.bitwise_and(img, img, mask=mask)

# Display the original image and segmented image side by side
cv2_imshow(np.hstack((img, res)))
```

```python
# Import necessary libraries
import cv2
from google.colab.patches import cv2_imshow

# Load the image
img = cv2.imread('crossword.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply binary thresholding with threshold value 200
_, thresh = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY)

# Display the original image and segmented image side by side
cv2_imshow(np.hstack((gray, thresh)))
```
